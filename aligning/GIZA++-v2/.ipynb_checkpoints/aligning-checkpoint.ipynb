```python
import re
import mysql.connector
from pprint import pprint
import pandas as pd
import os
from bs4 import BeautifulSoup

df = pd.read_csv("data.csv")

```

```python
def clean_text(text):
    text = BeautifulSoup(text).get_text(separator = " ")
    #text = re.sub("[^-a-zA-ZåäöÅÄÖ'éÉàÀ.,]", " ", text) 
    output = ""
    for char in text:
        if char == "." or char == ",":
            output += " " + char + " "
        else:
            output += char   
    return " ".join(output.lower().split())
```

```python

def prepare_texts(language):
    with open("{}.txt".format(language), "w") as file:
        for ind, row in df.iterrows():
            text = row["text_{}".format(language)]
            text = clean_text(text)
            file.write(text + "\n")
    os.system("./../mkcls-v2/mkcls -p{}.txt -V{}.txt.vcb.classes".format(language, language))

def align_texts(source, target):
    os.system("./plain2snt.out {}.txt {}.txt".format(source, target))
    os.system("./snt2cooc.out {}.vcb {}.vcb {}_{}.snt > corp_{}_{}.cooc".format(source, target,
                                                    source, target, source, target))
    if not os.path.exists("myout_{}_{}".format(source, target)):
        os.system("mkdir myout_{}_{}".format(source, target))
        
    cmd =  './GIZA++ -S {}.vcb -T {}.vcb -C {}_{}.snt -CoocurrenceFile corp_{}_{}.cooc --parallel -outputpath myout_{}_{} -o align'
    os.system(cmd.format(source, target, source, target, source, target, source, target))

def clean():
    os.system("make clean") 
    os.system("rm -rf myout*") 
    os.system("rm -f *vcb*")
    os.system("rm -f *snt")
    os.system("rm -f *cooc")
    os.system("make")

clean()

prepare_texts("sv")
prepare_texts("en")
prepare_texts("es")
prepare_texts("fr")

align_texts("sv", "en")
align_texts("fr", "en")
align_texts("es", "en")

```

```python
from nltk.stem import SnowballStemmer, WordNetLemmatizer
import spacy

stemmers = [WordNetLemmatizer(), SnowballStemmer("swedish"), spacy.load("fr_core_news_md"), spacy.load("es_core_news_md")]

def find_senses(amb_word, df_train):
    lemma = stemmers[0].lemmatize(amb_word.lower())
    senses_to_inds = dict()
    
    file_sv_en = open("myout_sv_en/align.VA3.final", "r")
    file_fr_en = open("myout_fr_en/align.VA3.final", "r")
    file_es_en = open("myout_es_en/align.VA3.final", "r")
        
    lang_lines = [file_sv_en.readlines(), file_fr_en.readlines(), file_es_en.readlines()]
    file_sv_en.close()
    file_fr_en.close()
    file_es_en.close()
    
    for ind, text in df.iterrows(): 
        en_text = lang_lines[0][3*ind+1].split()
        jacket_indices = [ind for ind, word in enumerate(en_text) if stemmers[0].lemmatize(word) == lemma]
        if len(jacket_indices) == 0:
            continue
            
        alignments = [lines[3*ind+2].split(" ({") for lines in lang_lines]        
        for jacket_index in jacket_indices:
            sense = [lemma]
            for lang_ind, alignment in enumerate(alignments):
                
                for i in range(1, len(alignment) - 1, 1):
                    sen1 = alignment[i]
                    sen2 = alignment[i+1]
                    break_ind2 = sen2.index("}")
                    break_ind1 = sen1.index("}")
                    
                    word = sen1[break_ind1+2:].strip()
                    inds = sen2[:break_ind2].split()

                    for ind in inds:
                        if jacket_index + 1 == int(ind):
                            sense += [word]
                            
            if len(sense) == 4:   
                sense = tuple(sense)
                df_train.loc[len(df_train)] = [" ".join(en_text), jacket_index, sense, -1]
                                                            
                if sense not in senses_to_inds:
                    senses_to_inds[sense] = set()
                senses_to_inds[sense].add(len(df_train)-1)
                
    return senses_to_inds

df_train = pd.DataFrame(columns = ["text", "ind", "alignment", "label"])
senses_to_inds = find_senses("jacket", df_train)
```

```python
from collections import Counter

def find_common_swedish_words(senses_to_inds, tot, threshold = 0.05):
    swedish_words = []
    for (_, sv_word, _, _), inds in senses_to_inds.items():
        swedish_words += [sv_word]*len(inds)
    
    counter_swedish_words = Counter(swedish_words)

    common_swedish_words = []
    for word in counter_swedish_words:
        count = counter_swedish_words[word]
        if count >= int(tot*threshold):
            common_swedish_words.append((word, count))

    common_swedish_words.sort(key = lambda tup: tup[1], reverse=True)
    return common_swedish_words

def merge_swedish_words(common_swedish_words, senses_to_inds):
    
    senses_to_inds_copy = senses_to_inds.copy()
    for sense in senses_to_inds.keys():
        en_word, sv_word, fr_word, es_word = sense
        for common_swedish_word, _ in common_swedish_words:

            if common_swedish_word in sv_word and common_swedish_word != sv_word:
                new_sense = tuple([en_word, common_swedish_word, fr_word, es_word])

                if sense in senses_to_inds_copy:
                    if new_sense not in senses_to_inds_copy:
                        senses_to_inds_copy[new_sense] = set()
                    senses_to_inds_copy[new_sense] = senses_to_inds_copy[new_sense].union(senses_to_inds_copy[sense])  
                    senses_to_inds_copy.pop(sense)
                break               
    return senses_to_inds_copy 

def merge_senses_stem(senses_to_inds): 
    
    senses_to_inds_copy = senses_to_inds.copy()
    for sense in senses_to_inds.keys():
        en_word, sv_word, fr_word, es_word = sense
        new_sense = (en_word, stemmers[1].stem(sv_word), 
                    stemmers[2][0](fr_word).lemma_, stemmers[3].stem(es_word)[0].token.lemma_)
        
        if sense in senses_to_inds_copy:
            if new_sense not in senses_to_inds_copy:
                senses_to_inds_copy[new_sense] = set()
            senses_to_inds_copy[new_sense] = senses_to_inds_copy[new_sense].union(senses_to_inds_copy[sense])  
            senses_to_inds_copy.pop(sense)
        
    return senses_to_inds_copy 

def cut_off_senses(senses_to_inds, tot, threshold = 0.05):
    
    senses_to_inds_copy = senses_to_inds.copy()
    for sense, inds in senses_to_inds.items():
        if len(inds) < int(tot*threshold):
            senses_to_inds_copy.pop(sense)         
    return senses_to_inds_copy

def label_data(senses_to_inds, df_train):
    df_train = df_train.copy(deep = True)
    for label, (_, inds),  in enumerate(senses_to_inds.items()):
        df_train.loc[inds, "label"] = label
    return df_train.loc[df_train["label"] != -1]
        
        
tot = sum([len(inds) for _, inds in senses_to_inds.items()])

common_swedish_words = find_common_swedish_words(senses_to_inds, tot)
senses_to_inds = merge_swedish_words(common_swedish_words, senses_to_inds)

senses_to_inds = merge_senses_stem(senses_to_inds)
senses_to_inds = cut_off_senses(senses_to_inds, tot, threshold = 0.10)

df_train = label_data(senses_to_inds, df_train)

senses_to_count = dict.fromkeys(set(senses_to_inds))
for key, val in senses_to_inds.items():
    senses_to_count[key] = len(val)
    representations = Counter(df_train.loc[val, "alignment"])
    representation = representations.most_common(1)[0][0]
    en_word, sv_word, fr_word, es_word = representation
    for common_swedish_word,_ in common_swedish_words:
        if common_swedish_word in sv_word and common_swedish_word != sv_word:
            representation = (en_word, common_swedish_word, fr_word, es_word)
    senses_to_count[representation] = senses_to_count.pop(key)


```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df_sense = pd.DataFrame.from_dict(senses_to_count, orient = "index").reset_index()
df_sense.rename(columns = {"index" : "sense", 0 : "count"}, inplace=True)

fig = plt.figure(figsize = (6,7))
sns.barplot(x = "count", y = "sense", data = df_sense)
```

```python
def generate_training_data(word, PATH, save_hist = True, do_cancel = True):
    
    df_train = pd.DataFrame(columns = ["text", "ind", "alignment", "label"])
    senses_to_inds = find_senses(word, df_train)

    tot = sum([len(inds) for _, inds in senses_to_inds.items()])

    common_swedish_words = find_common_swedish_words(senses_to_inds, tot)
    senses_to_inds = merge_swedish_words(common_swedish_words, senses_to_inds)

    senses_to_inds = merge_senses_stem(senses_to_inds)
    senses_to_inds = cut_off_senses(senses_to_inds, tot, threshold = 0.10)
    
    if do_cancel:
        if len(senses_to_inds.keys()) < 2:
            return None

    df_train = label_data(senses_to_inds, df_train)

    senses_to_count = dict.fromkeys(set(senses_to_inds))
    for key, val in senses_to_inds.items():
        senses_to_count[key] = len(val)
        representations = Counter(df_train.loc[val, "alignment"])
        representation = representations.most_common(1)[0][0]
        en_word, sv_word, fr_word, es_word = representation
        for common_swedish_word,_ in common_swedish_words:
            if common_swedish_word in sv_word and common_swedish_word != sv_word:
                representation = (en_word, common_swedish_word, fr_word, es_word)
        senses_to_count[representation] = senses_to_count.pop(key)

    df_sense = pd.DataFrame.from_dict(senses_to_count, orient = "index").reset_index()
    df_sense.rename(columns = {"index" : "sense", 0 : "count"}, inplace=True)
    
    if save_hist:
        plt.ioff()
        fig = plt.figure(figsize = (6,7))
        sns.barplot(x = "count", y = "sense", data = df_sense)
        plt.savefig(PATH + word + ".png",bbox_inches='tight')
        plt.close()
    else:
        fig = plt.figure(figsize = (6,7))
        sns.barplot(x = "count", y = "sense", data = df_sense)
    
    return df_train
            
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

words = df["text_en"].apply(lambda text: clean_text(text).split()).values
words = Counter([lemmatizer.lemmatize(word) for sublist in words for word in sublist if word not in stop_words])
words = set([key for key in words.keys() if words[key] >= 150 ])
PATH = "../../data/train/"

for word in tqdm(words):
    df_train = generate_training_data(word, PATH, save_hist=True, do_cancel=True)
    if df_train is not None:
        df_train.to_csv(PATH + word + ".csv", index=False)

```
